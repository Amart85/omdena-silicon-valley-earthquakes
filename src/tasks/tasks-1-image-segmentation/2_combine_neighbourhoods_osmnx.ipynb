```python
import osmnx as ox
import geopandas
import matplotlib.pyplot as plt
```

```python

# northridge
G_northridge_all   = ox.graph_from_place('Northridge, California, USA')
G_northridge_nodes,G_northridge_edges  = ox.graph_to_gdfs(G_northridge_all)

# north hills
G_nh_all   = ox.graph_from_place('North Hills, California, USA')
G_nh_nodes, G_nh_edges  = ox.graph_to_gdfs(G_nh_all)

# combine them
G_all_nodes = G_northridge_nodes.append(G_nh_nodes)
G_all_edges = G_northridge_edges.append(G_nh_edges)

# fetch new graph
graph_all = ox.graph_from_gdfs(G_all_nodes,G_all_edges)
new_nodes,new_edges = ox.graph_to_gdfs(graph_all)
fig, ax = plt.subplots(figsize=(10, 10))
```

```python

fig, (ax1,ax2,ax3) = plt.subplots(1,3,figsize=(20,20))
G_northridge_edges.plot(ax=ax1, linewidth=1,edgecolor='#BC8F8F')

G_nh_edges.plot(ax=ax2, linewidth=1,edgecolor='#BC8F8F')

new_edges.plot(ax=ax3, linewidth=1,edgecolor='#BC8F8F')
fig.tight_layout()
```
